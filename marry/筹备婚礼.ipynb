## 朋友婚礼有感

## witness 

参加了一个很要好的朋友的婚礼之后,很高兴亲自见证了朋友的婚礼,也算完满了.

很多人不能理解我们这些活跃在github的中国人是为了什么?每次都是积极的关注技术的发展.然而不知不觉忘记了生活的意义和价值.

作为一个纯粹的技术人,该露一手的时候还是要出手的.特别是男生,参加了这次婚礼,我意识到了很多自己将来要面对的问题.

因为长时间生活在第三世界.我就能清楚的感觉到.自己和第一世界的距离越来越远.哪怕我是80后也好.可惜不是.

我已经和父辈们的认知已经有很大的不同.但是要做到父辈满意的程度,这真的是很难很难.

但是你又不能把脑袋里面的东西给倒出来.让别人瞅瞅.解释给别人听,他又听不懂.

有些事,你不做,会遗憾的.

参加婚礼之前,我真的没想过结婚的事.

参加婚礼之后,我觉得我不想抵触了(kill shadow).


## 苦 & 衷

有一个哥们,跟他女朋友,我们关系不错.然后很狗血的只有女的参加了婚礼.

我们都是高中时代一起走过来的朋友.而且都是当地的,可以说他们的感情很真挚了.

我跟她只有一面之缘,但是让人家女生说出`"不想再耽误通子了"`,`"感觉我爱他比他爱的多"`,`"他想在南方,我想在家这边"`.这些话,

我也不知道怎么和女方说好话了.

有爱而无所得,情深而互相伤痛.

连我这种单身很久的人,也会为此惋惜.

另外,我真的不知道为什么一个女的会被逼到这种地步.

一个女生先提出分手.那还有个J8意思.

爱情来的时候你不知道,爱情走的时候你又不知道.

其实大家心里明白.无疾而终的爱情收获的不仅仅是回忆.更是教训,不要等问题出现的时候,再去弥补.

而是用心维护,防患于未然.

不说时刻提醒自己吧.但是请记住.你妈永远是你妈,女朋友不一定会是你的妻子.

## 思 & 虑

爱情需要机缘.婚礼需要长期的设计.

我去了三天:

第一天: 布置家里婚房.

第二天: 布置酒店的婚房.

第三天: 参加婚礼.

他们家里是真的没人有创意.所以只能求个圆满.而不是好上加好.

## 打 & 算

技术是用来创造幸福的,而不是一个冰冷的统治工具.

很多话,我作为一个开发者,并不能说,因为说了,中国的一些社会现象也并不能改变.

但是我所做的一切,真的只是为了说明一个事: 码农也懂得浪漫.

有个大爷姐说过程序员兄弟很容易被绿.他单位的一个女怎么怎么的.

总以为我们是个呆瓜.也不知道我们整天做什么.

只是作为程序员,不应该只关注于技术,还应该离自己的亲人和爱人近一些.

我就想用实际行动的告诉这些目光短浅的傻逼.

你的程序员爸爸不但没有社会戾气,而且很有朝气.

他们热爱生活,懂得制造快乐.懂浪漫.顾家.就是比尼玛的2B营销做事光明磊落.

不服,就进食物链.麻痹的设计更复杂的系统难死你个沙比.
