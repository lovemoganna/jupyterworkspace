## Array map & fill 

`map(function(){})` --  map里面必须是一个函数,因为他会返回一个新的数组.

`fill(n)` -- 以n来填充这个数组,Array(100)的值是`empty`,`Array(100).fill()`以空值填充,所以相当于声明了数组中的100个变量.

```js
// map的测试
var arr = [7.8, 4, 3, 2]

console.log(arr);

function double(val) {
  return val * 2
}

// map会返回一个新的数组
// map里面必须是一个函数,因为它需要返回一个新的数组.

// var doubler = arr.map(double)
arr = arr.map(double)

console.log(arr);

// 箭头函数改造

arr = arr.map(x => x * 2);

console.log(arr);


// fill的测试

// 以...填充,生成4个相同的数字.

arr.fill(0); // [0,0,0,0]

console.log(arr)

// 四舍五入只要保留2位数字.
arr.fill((Math.random() * 1).toFixed(2))

console.log(arr)


let arr1 = new Array(100)

console.log(arr1);

// 生成100个数
arr1 = arr1.fill(0).map(x => (Math.random() * 10).toFixed(2))
console.log(arr1);

// cause Math.random is a function
var arr2 = arr1.fill(0).map(Math.random)
console.log(arr2);

// Array(100).fill() 声明100个空数组.
var vals = Array(100).fill().map(Math.random)
console.log(vals);

```

## Array的reduce

主要是来处理层层遍历的.里面有2个参数,第一个参数是层层遍历后的结果,第二个参数是遍历前的初始值.

```js
let vals = [5, 4, 3, 2, 9]

// let acc = 0;

// for (const val of vals){
//   acc += val
// }
// console.log(acc);


let sum = function (acc, val) {
  console.log(acc);
  return acc + val;
}
// 定义sum的初始值
// 一直调用sum和这个函数.直至遍历vals完毕.
let answer = vals.reduce(sum, 10);
console.log(answer);


// the sum initial value is 10 

let answers = vals.reduce((acc, val) => {

  console.log('the result is ', acc, 'the value is ', val);
  return acc + val; // 返回的值充当第二轮的acc
},10)

console.log('hi multiply ' + answers)
```

### 比较大小
```js
let val2 = [5, 4, 3, 2, 9]

var getMax = function (n1, n2) {
  // if (n1 > n2) {
  //   return n1
  // }else{
  //   return n2
  // }
 
  // 这两种写法都可以
  
  if (n1 > n2) {
    n2 = n1
  }
  return n2;
}


var r = val2.reduce(getMax)
console.log(r);
```

```js
var r = val2.reduce((n1, n2) => {
  console.log('比较大的值是 ',n1,' 下一个送死的是 ',n2);
  return n2 > n1 ? n2 : n1
},8)
console.log(r);
```

## Array Sort

```js
// sort不会比较键值对,但是可以比较他们的值
let vals = [{
  x: 5,
  y: 6
}, {
  x: 2,
  y: 10
}]

function compare(a, b) {
  // 按着y的值来比较.
  return b.y - a.y;
}

var result = vals.sort(compare) // 不起租用

console.log(result);


let valsTwo = ['a', 'hello', 'b', 'B', 'byebye', 5, 9, 'A']
// [5, 9, "A", "B", "a", "b", "byebye", "hello"]
var re = valsTwo.sort()

console.log(re);

let s = "it was  a dark and stroy ninght.";

let words = s.split(/\W+/).filter(word => word.length >= 3) // 检测出数组中字符串的长度大于3的值

// 也可以根据字符串的长度来比较

words.sort((a,b)=> b.length-a.length); // ["ninght", "stroy", "dark", "was", "and"]

console.log(words);
```
