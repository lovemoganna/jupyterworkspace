## 1-2-vue-todoList

```html
<!DOCTYPE html>
<html lang="en">

<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <meta http-equiv="X-UA-Compatible" content="ie=edge">
    <script src="https://unpkg.com/vue"></script>
    <title>Document</title>
</head>

<body>

    <div id="app">
        <!-- v-modal 数据的双向绑定 
             在此绑定的是 inputValue这个值.
             也就是当输入框发生改变的时候,vue实例中的inputValue会发生改变.
             同样,当data中的inputValue值发生改变,输入框中的值也会发生改变.

             我们在控制台打印的结果:
             app.$data.inputValue
             默认值是""

             如果此时在下面输入 `app.$data.inputValue === '1234'`,输入框里面也会是1234
             同样,你在输入框输入`1234`,此时的app.$data.inputValue打印出来的值也是 1234.
             这就是数据双向绑定的示例.


    
        -->
        <input type="text" v-model="inputValue">
        <!-- 绑定事件 -->
        <button v-on:click="handleBtnClick">提交</button>

        <ul>
            <!-- 遍历TODOList -->
            <li v-for="(item,index) of list">{{item}}</li>
        </ul>
    </div>

    <script>

        var app = new Vue({
            el: '#app',
            data() {
                return {
                    list: ['第一', '第二', '第三'],
                    inputValue: ''
                }
            },
            methods: {
                handleBtnClick: function () {
                    // 打印输入框的内容
                    alert(this.inputValue)
                    this.list.push(this.inputValue)

                    // 去除掉原始值
                    this.inputValue = ''
                }
            }
        })

        /*
            从头到尾我们都在操纵数据.

            todolist的数据我们存放在 List这个数组当中.

            input输入框的数据我们存放在 inputValue这个对象当中.

            所以我们直接改变的是数据.然后页面就自动的发生了变化.

            注意我们现在跟DOM的操纵没有任何关系了.

            这种模式就是 MVVM模式.

            也就是说我们将DOM层的操作改为了数据层的操作.

            vue的底层会自动的根据数据的不同给我们自动的去渲染页面.
        **/

    </script>
</body>
```

## 1-3-mvvm model

### 传统的MVP模式

- model 数据层(ajax获取数据)
- Presenter 呈现层(业务逻辑)
- view DOM展示(视图)

视图层发生改变(触发一个事件),控制器中的代码会去执行.
执行ajax,会执行数据层.

业务逻辑(Presenter)中大部分会在操作DOM.

![](https://upload-images.jianshu.io/upload_images/7505161-b8d4391ae582ebd9.png?imageMogr2/auto-orient/strip%7CimageView2/2/w/1240)


```html
<!DOCTYPE html>
<html lang="en">

<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <meta http-equiv="X-UA-Compatible" content="ie=edge">
    <!-- <script src="https://unpkg.com/vue"></script> -->
    <script src="https://unpkg.com/jquery"></script>

    <title>Document</title>
</head>

<body>

    <div>
        <input type="text" id="inputValue">
        <button id='btn'>提交</button>
        <ul id='list'>
        </ul>
    </div>
    <script>
        function Page(){
        }
        $.extend(Page.prototype,{
            init: function(){
                // 初始化绑定事件
                this.bindEvents()
            },
            bindEvents: function(){
                // 详细处理
                var btn = $('#btn')
                // $.proxy()改变this的指向
                // 将handleBtnClick这个函数绑定成Page实例的引用
                btn.on('click',$.proxy(this.handleBtnClick,this))
            },
            handleBtnClick: function(){
                var valueObj = $('#inputValue')
                
                var value = valueObj.val()
                
                var ulEle = $('#list')
                
                ulEle.append('<li>'+value+'</li>')
                
                valueObj.val(' ')
            }
        })

        var page = new Page()
        page.init()
    </script>
    
</body>
```

## 改造TodoList

### 全局组件

```html
<!DOCTYPE html>
<html lang="en">

<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <meta http-equiv="X-UA-Compatible" content="ie=edge">
    <script src="https://unpkg.com/vue"></script>
    <title>Document</title>
</head>

<body>

    <div id="app">

        <input type="text" v-model="inputValue">
        <!-- 绑定事件 -->
        <button v-on:click="handleBtnClick">提交</button>

        <ul>
            <!-- 引用全局组件
                v-for决定了使用这个组件的次数
                借助,content这个参数来传递 item的值.
            -->
            <todo-item :content="item" v-for="(item,index) of list" :key="item.id">{{item}}</todo-item>
        </ul>
    </div>

    <script>
        // 创建全局组件
        Vue.component('TodoItem', {
            
            // 接收父组件传过来的参数
            props: ['content'],
            template: "<li>{{content}}</li>"
        })
        var app = new Vue({
            el: '#app',
            data() {
                return {
                    list: ['第一', '第二', '第三'],
                    inputValue: ''
                }
            },
            methods: {
                handleBtnClick: function () {
                    this.list.push(this.inputValue)
                    // 去除掉原始值
                    this.inputValue = ''
                }
            }
        })
    </script>
</body>
```

## 组件中的传值

### 父组件向子组件中传值

```html
<!DOCTYPE html>
<html lang="en">

<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <meta http-equiv="X-UA-Compatible" content="ie=edge">
    <script src="https://unpkg.com/vue"></script>
    <title>Document</title>
</head>

<body>

    <div id="app">

        <input type="text" v-model="inputValue">
        <!-- 绑定事件 -->
        <button v-on:click="handleBtnClick">提交</button>

        <ul>
            <!-- 引用全局组件
                v-for决定了使用这个组件的次数
                借助,content这个参数来传递 item的值.
            -->
            <todo-item :content="item" v-for="(item,index) of list" :key="item.id">{{item}}</todo-item>
        </ul>
    </div>

    <script>
        // 局部组件,需要注册到vue实例中
        var TodoItem = {
            props: ['content'],
            template: '<li>{{content}}</li>'

        }
        var app = new Vue({
            el: '#app',
            components: {
                TodoItem
            },
            data() {
                return {
                    list: ['第一', '第二', '第三'],
                    inputValue: ''
                }
            },
            methods: {
                handleBtnClick: function () {
                    this.list.push(this.inputValue)
                    // 去除掉原始值
                    this.inputValue = ''
                }
            }
        })

        
    </script>
</body>
```

- 父组件向子组件传值的方式是 通过 `v-bind:content`的形式来进行传值.

- 子组件通过`props:['content']`来接收值,并利用插值表达式来使用这个值.

##  子组件给父组件传值

- 子组件向父组件传值的方式是,在子组件里面通过`this.$emit`绑定一个事件和参数(参数需要在`props`里面声明))

- 进而绑定父组件,让父组件来监听这个事件.

- 然后去Vue实例中去详细写这个事件的具体步骤,同时也可以接收到参数,这样子组件就传给了父组件值.

```html
<!DOCTYPE html>
<html lang="en">

<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <meta http-equiv="X-UA-Compatible" content="ie=edge">
    <script src="https://unpkg.com/vue"></script>
    <title>Document</title>
</head>

<body>

    <div id="app">

        <input type="text" v-model="inputValue">
        <button v-on:click="handleBtnClick">提交</button>

        <ul>
            <todo-item :content="item" v-for="(item,index) of list" :key="index" @delete="handleItemDelete">{{item}}</todo-item>
            <!-- handleItemDelete事件需要在vue实例中声明 -->
        </ul>
    </div>

    <script>
        // 局部组件,需要注册到vue实例中
        var TodoItem = {
            props: ['content', 'index'],
            // 这里给子组件新添加了一个方法,需要写在子组件当中.
            template: '<li @click="handleItemClick">{{content}}</li>',
            methods: {
                handleItemClick: function () {
                    // this.$emit() 向外触发一个事件!
                    this.$emit('delete', this.index)
                }
            }
        }
        var app = new Vue({
            el: '#app',
            components: {
                TodoItem
            },
            data() {
                return {
                    list: [],
                    inputValue: ''
                }
            },
            methods: {
                handleBtnClick: function () {
                    this.list.push(this.inputValue)
                    // 去除掉原始值
                    this.inputValue = ''
                },
                handleItemDelete: function (index) {
                    // 从下标的位置开始删除
                    this.list.reverse().splice(index,1)
                }
            }
        })
    </script>
</body>
```
